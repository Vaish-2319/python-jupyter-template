### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Data points
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create scatter plot
plt.scatter(x, y, color='blue', marker='o')

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=12, ha='right', va='bottom')

# Customize the plot
plt.title('Scatter Plot with Annotations')  # Add title
plt.xlabel('X values')  # X-axis label
plt.ylabel('Y values')  # Y-axis label
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid

# Show the plot
plt.show()
```
