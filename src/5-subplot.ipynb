### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Creating a figure with a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot (First subplot)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Line Chart")

# Bar Chart (Second subplot)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axes[0, 1].bar(categories, values, color=['blue', 'red', 'green', 'orange'])
axes[0, 1].set_title("Bar Chart")

# Scatter Plot (Third subplot)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x, y, color='purple')
axes[1, 0].set_title("Scatter Plot")

# Pie Chart (Fourth subplot)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()
```
