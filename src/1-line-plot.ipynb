### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values from 0 to 10
x = np.arange(0, 11)  # Generates numbers from 0 to 10
y = x**2  # Compute y as the square of x

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')

# Customize the plot
plt.title('Line Plot: y = x²')  # Add title
plt.xlabel('X values')  # Label for X-axis
plt.ylabel('Y values')  # Label for Y-axis
plt.grid(True)  # Enable grid
plt.legend()  # Add legend

# Show the plot
plt.show()
```
